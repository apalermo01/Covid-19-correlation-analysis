# Analyze regression results (r-squared only)

# 0. Imports

```python
from covid_project.regression_funcs import collect_all_regression_results_to_df, plot_rsquared_heatmap
```

# 1. Generate visualization

```python
path = "./data/regression_results/"


df = collect_all_regression_results_to_df(path)
dep_vars = [
    'new_cases_1e6',
    'new_cases_7day_1e6',
    'new_deaths_1e6',
    'new_deaths_7day_1e6'
]

for var in dep_vars:
    _, bins_ids = plot_rsquared_heatmap(
        data = df,
        dep_var = var,
        sort_values = True,
        ax = None
    )
```
